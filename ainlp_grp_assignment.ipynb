```python
from generation.generate import generate_caption
from generation.image_generator import MemeImageGenerator
from rendering.render import render_meme
from PIL import Image
from IPython.display import display

# 1. Generate caption
caption = generate_caption(
    topic="monday mornings",
    style="genz",
    intensity=0.8,
    temperature=0.8,
    top_p=0.9
)

print("Caption:", caption)

# 2. Generate image
gen = MemeImageGenerator()
raw_image_path = gen.generate(
    topic="monday mornings",
    style="genz",
    intensity=0.9
)

# Optional: show raw image
display(Image.open(raw_image_path))

# 3. Render final meme (clean + caption)
final_image_path = render_meme(
    caption=caption,
    image_path=raw_image_path
)

# 4. Display final meme
display(Image.open(final_image_path))
```
